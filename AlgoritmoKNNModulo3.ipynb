# Algoritmo KNN (K-nearest neighbors)

O KNN tenta classificar cada amostra de um conjunto de dados avaliando sua distância em relação aos vizinhos mais próximos.

## Importando Dataset Iris

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

## Gerar um target (Alvo)

```python
df['target']=iris.target
df
```

## Gerar um target name (apenas para demonstrar)

```python
iris.target_names
```

```python
df['target_name']=iris.target_names[df['target']]
```

```python
df
```

## Determinar as features (colunas, atributos) que vão fazer parte do estudo

```python
df.columns
```

```python
iris_features=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']
```

##Determinar nosso alvo (y), lembrando que por fazer uma classificação usando KNN precisamos de um alvo numérico

```python
y=df.target
```

##Selecionar nossos dados, as features que serão utilizadas (X)

Costuma-se colocar dados/features em maísculo "X" e o alvo em minúsculo "x".

```python
X=df[iris_features]
```

```python
X
```

```python
y
```

## Usar Scikit-Learn

* Define = Escolha de modelo (define parâmetro)
* Fit = Treinar
* Predict = Fazer a predição
* Evaluate = Avaliar os resultados

## Importar o algoritmo KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

A importação será com dados de classificação, pois a partir dos dados queremos que ele nos diga qual tipo de flor é.

## Selecionar o modelo

```python
modelo = KNeighborsClassifier(3)
```

## Treino

No treino eu vou pegar as features que eu tenho e os resultados que eu tenho (tabela original) e vou mandar ele treinar.

```python
modelo.fit(X,y)
```

```python
X
```

## Ver a predição

Para cada X o algoritmo vai gerar o y, baseado no modelo que ele treinou.

```python
print(modelo.predict(X))
```

Essa predição tem alguns erros, não está conseguindo bater exatamente com os resultados que eu tinha.

## Vamos observar a Acurácia (porcentagem de acertos), lembrando que estamos colocando pra validação os mesmos dados que foram treinados.

```python
modelo.score(X,y)
```

Portanto, o modelo conseguiu capturar 96%, errando apenas 4%.
